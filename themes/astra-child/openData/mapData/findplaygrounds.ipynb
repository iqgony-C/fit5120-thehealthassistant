### Create a map for data

```python

m = folium.Map(location=[-38.043995, 145.264296], control_scale=True, zoom_start=6,attr = "text some")
df1_copy = df1.copy()
df2_copy = df2.copy()
df3_copy = df3.copy()
df4_copy = df4.copy()
df5_copy = df5.copy()
df6_copy = df6.copy()
df7_copy = df7.copy()
df8_copy = df8.copy()
df9_copy = df9.copy()
df10_copy = df10.copy()
df11_copy = df11.copy()
df12_copy = df12.copy()
df13_copy = df13.copy()
fg = folium.FeatureGroup()    

# loop through data to create Marker for each hospital
for i in range(0,len(df1_copy)):
    
    
    folium.Marker(
    location=[df1_copy.iloc[i]['LATITUDE'], df1_copy.iloc[i]['LONGITUDE']],
    #popup="Playground",
    tooltip=str(df1_copy.iloc[i]['LOCATION']),
    icon=folium.Icon(color='blue',icon='info-sign'),
    ).add_to(fg)
   

for i in range(0,len(df2_copy)):
    
    
    folium.Marker(
    location=[df2_copy.iloc[i]['lat'], df2_copy.iloc[i]['long']],
    #popup="Playground",
    tooltip=str(df2_copy.iloc[i]['Area']),
    icon=folium.Icon(color='blue',icon='info-sign'),
    ).add_to(fg)    
    
for i in range(0,len(df3_copy)):
    
    
    folium.Marker(
    location=[df3_copy.iloc[i]['lat'], df3_copy.iloc[i]['long']],
    #popup="Playground",
    tooltip=str(df3_copy.iloc[i]['name']),
    icon=folium.Icon(color='blue',icon='info-sign'),
    ).add_to(fg)   


for i in range(0,len(df4_copy)):
    
    
    folium.Marker(
    location=[df4_copy.iloc[i]['lat'], df4_copy.iloc[i]['long']],
    #popup="Playground",
    tooltip=str(df4_copy.iloc[i]['name']),
    icon=folium.Icon(color='blue',icon='info-sign'),
    ).add_to(fg)   


for i in range(0,len(df5_copy)):
    
    
    folium.Marker(
    location=[df5_copy.iloc[i]['Latitude'], df5_copy.iloc[i]['Longitude']],
    #popup="Playground",
    tooltip=str(df5_copy.iloc[i]['Name']),
    icon=folium.Icon(color='blue',icon='info-sign'),
    ).add_to(fg)
    
for i in range(0,len(df6_copy)):
    
    
    folium.Marker(
    location=[df6_copy.iloc[i]['lat'], df6_copy.iloc[i]['lon']],
    #popup="Playground",
    tooltip=str(df6_copy.iloc[i]['name']),
    icon=folium.Icon(color='blue',icon='info-sign'),
    ).add_to(fg)     
    
for i in range(0,len(df7_copy)):
    
    
    folium.Marker(
    location=[df7_copy.iloc[i]['LATITUDE'], df7_copy.iloc[i]['LONGITUDE']],
    #popup="Playground",
    tooltip=str(df7_copy.iloc[i]['PARK_NAME']),
    icon=folium.Icon(color='blue',icon='info-sign'),
    ).add_to(fg)
    
for i in range(0,len(df8_copy)):
    
    
    folium.Marker(
    location=[df8_copy.iloc[i]['Latitude'], df8_copy.iloc[i]['Longitude']],
    #popup="Playground",
    tooltip=str(df8_copy.iloc[i]['Name']),
    icon=folium.Icon(color='blue',icon='info-sign'),
    ).add_to(fg)

for i in range(0,len(df9_copy)):
    
    
    folium.Marker(
    location=[df9_copy.iloc[i]['Latitude'], df9_copy.iloc[i]['Longitude']],
    #popup="Playground",
    tooltip=str(df9_copy.iloc[i]['Name']),
    icon=folium.Icon(color='blue',icon='info-sign'),
    ).add_to(fg)
    
for i in range(0,len(df10_copy)):
    
    
    folium.Marker(
    location=[df10_copy.iloc[i]['Lat'], df10_copy.iloc[i]['Long']],
    #popup="Playground",
    tooltip=str(df10_copy.iloc[i]['PLAY_GROUND']),
    icon=folium.Icon(color='blue',icon='info-sign'),
    ).add_to(fg)
    
for i in range(0,len(df11_copy)):
    
    
    folium.Marker(
    location=[df11_copy.iloc[i]['Latitude'], df11_copy.iloc[i]['Longitude']],
    #popup="Playground",
    tooltip=str(df11_copy.iloc[i]['Name']),
    icon=folium.Icon(color='blue',icon='info-sign'),
    ).add_to(fg)
    
for i in range(0,len(df12_copy)):
    
    
    folium.Marker(
    location=[df12_copy.iloc[i]['LAT'], df12_copy.iloc[i]['LONG']],
    #popup="Playground",
    tooltip=str(df12_copy.iloc[i]['Complex_Name']),
    icon=folium.Icon(color='blue',icon='info-sign'),
    ).add_to(fg)   

for i in range(0,len(df13_copy)):
    
    
    folium.Marker(
    location=[df13_copy.iloc[i]['lat'], df13_copy.iloc[i]['lon']],
    #popup="Playground",
    tooltip=str(df13_copy.iloc[i]['name']),
    icon=folium.Icon(color='blue',icon='info-sign'),
    ).add_to(fg)
    
    
    
fg.add_to(m)
#Search(fg).add_to(m)


m
```
